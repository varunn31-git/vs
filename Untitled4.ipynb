```python
import matplotlib.pyplot as plt

data = [5,6,7,7,8,9,10,10,10,11,12,11,11]

plt.hist(data,bins= 10, edgecolor='black')

plt.xlabel('Numbers')

plt.ylabel('Count')

plt.title('Histogram of Data')

plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

datf = pd.DataFrame({"Season 1":[2,4,6,8,3],"Season 2":[1,4,6,4,3]})

p = sns.histplot(data=datf)

p.set(xlabel="X Label Value",ylabel = "y Label Value")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randint(20,81,1000)

plt.hist(data,bins= 15, edgecolor='black', color='skyblue')

plt.title('Hisrogram of Cancer Patients Age Distribution')

plt.xlabel('Age')
plt.ylabel('Number of Patients')


plt.show()
```
